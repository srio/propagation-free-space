```python
import numpy as np
import matplotlib.pyplot as plt
from xwp.spectral_2d import propTF
```

```python
energy = 10000                     #Energy in EV
wavel = (1239.84/energy)*10**(-9)  #Wavelength in meters
z = 75e-6
N = 2048
L_in  = 5e-6
```

```python
x_in = np.linspace(-L_in/2,L_in/2,N)
y_in = np.linspace(-L_in/2,L_in/2,N)
X,Y = np.meshgrid(x_in,y_in)
in_coords = np.array([-L_in/2,L_in/2,-L_in/2,L_in/2])*1e6
```

```python
aperture = np.zeros((N,N))
aperture[np.where((np.abs(X)<0.625e-6) & (np.abs(Y)<0.625e-6))] = 1
in_wave = np.ones((N,N))
in_wave = in_wave*aperture
del aperture
```

```python
plt.imshow(in_wave, extent = in_coords )
plt.colorbar()
plt.show()
```

```python
out_wave,L_out = propTF(in_wave,L_in/N,L_in,wavel,z)
x_out = np.linspace(-L_out/2,L_out/2,N)
y_out = np.linspace(-L_out/2,L_out/2,N)
out_coords = np.array([-L_out/2,L_out/2,-L_out/2,L_out/2])*1e6
```

```python
plt.rcParams["figure.figsize"] = (8,8)
f,(ax1,ax2) = plt.subplots(1,2)
ax1.imshow(np.abs(out_wave),extent=out_coords)
ax2.plot(x_out*1e6,np.abs(out_wave[int(N/2),:])) 
f.subplots_adjust(right=1.75)
plt.show()
```
